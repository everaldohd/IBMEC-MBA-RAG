## 📦 CÉLULA 3 — Configurar Variáveis de Ambiente com Segurança

**O que faz:** Configura chaves de API e variáveis de ambiente de forma segura usando `google.colab.userdata`.

**Por que:** **NUNCA** cole chaves de API diretamente no código — o notebook pode ser compartilhado por acidente. O Colab oferece um cofre seguro de segredos acessível via `userdata`.

> **Como adicionar segredos no Colab:**
> 1. Clique no ícone 🔑 no painel esquerdo ("Segredos")
> 2. Clique em "+ Adicionar novo segredo"
> 3. Nome: `LANGFUSE_PUBLIC_KEY`, Valor: (sua chave)
> 4. Ative o toggle "Acesso ao notebook"

````python
# Célula 3: Configurar variáveis de ambiente com segurança
import os

def get_secret_safe(nome, default='NAO_CONFIGURADO'):
    """Obtém segredo do Colab ou variável de ambiente, com fallback."""
    # Tenta obter do cofre do Colab
    try:
        from google.colab import userdata
        valor = userdata.get(nome)
        if valor:
            return valor
    except Exception:
        pass
    
    # Tenta variável de ambiente
    valor = os.environ.get(nome, default)
    return valor

# ─── Configurações do OpenSearch ─────────────────────────────
OPENSEARCH_HOST = os.environ.get('OPENSEARCH_HOST', 'localhost')
OPENSEARCH_PORT = int(os.environ.get('OPENSEARCH_PORT', '9200'))
OPENSEARCH_URL  = f'http://{OPENSEARCH_HOST}:{OPENSEARCH_PORT}'

# ─── Configurações do vLLM (LLM local) ────────────────
VLLM_HOST = os.environ.get('VLLM_HOST', 'localhost')
VLLM_PORT = int(os.environ.get('VLLM_PORT', '11434'))
VLLM_URL  = f'http://{VLLM_HOST}:{VLLM_PORT}'

# ─── LangFuse (Observabilidade) ───────────────────────────────
LANGFUSE_PUBLIC_KEY  = get_secret_safe('LANGFUSE_PUBLIC_KEY')
LANGFUSE_SECRET_KEY  = get_secret_safe('LANGFUSE_SECRET_KEY')
LANGFUSE_HOST        = get_secret_safe('LANGFUSE_HOST', 'https://cloud.langfuse.com')

# ─── Configura variáveis de ambiente para frameworks ─────────
os.environ['OPENSEARCH_URL']      = OPENSEARCH_URL
os.environ['VLLM_BASE_URL']     = VLLM_URL
os.environ['LANGFUSE_PUBLIC_KEY'] = LANGFUSE_PUBLIC_KEY
os.environ['LANGFUSE_SECRET_KEY'] = LANGFUSE_SECRET_KEY
os.environ['LANGFUSE_HOST']       = LANGFUSE_HOST

# ─── Exibe status (SEM revelar valores de chaves) ─────────────
print('⚙️  CONFIGURAÇÃO DO AMBIENTE')
print('=' * 60)
print(f'OpenSearch URL:      {OPENSEARCH_URL}')
print(f'vLLM URL:          {VLLM_URL}')
print(f'LangFuse Host:       {LANGFUSE_HOST}')

# Mostra apenas se a chave está configurada, NÃO o valor
def status_chave(nome, valor):
    if valor and valor != 'NAO_CONFIGURADO':
        return f'✅ Configurada ({len(valor)} chars)'
    return '⚠️  NÃO CONFIGURADA (configure nos Segredos do Colab)'

print(f'LangFuse Public Key: {status_chave("LANGFUSE_PUBLIC_KEY", LANGFUSE_PUBLIC_KEY)}')
print(f'LangFuse Secret Key: {status_chave("LANGFUSE_SECRET_KEY", LANGFUSE_SECRET_KEY)}')

print('\n✅ Variáveis de ambiente configuradas!')
````

## 📦 CÉLULA 4 — Validar Instalação de Todas as Bibliotecas

**O que faz:** Importa todas as bibliotecas instaladas e verifica suas versões.

**Por que:** É fundamental verificar que todas as importações funcionam antes de começar a trabalhar, pois conflitos de versão podem causar erros crypticos mais tarde.

````python
# Célula 4: Validar todas as importações

print('🔍 VALIDANDO INSTALAÇÕES...')
print('=' * 60)

resultados = []

def testar_import(modulo, nome_display=None, atributo_versao='__version__'):
    """Testa importação de um módulo e retorna status."""
    display = nome_display or modulo
    try:
        mod = __import__(modulo)
        versao = getattr(mod, atributo_versao, 'N/A')
        resultados.append((display, '✅', str(versao)))
    except ImportError as e:
        resultados.append((display, '❌', str(e)[:50]))

# Testa cada biblioteca
testar_import('sentence_transformers', 'sentence-transformers')
testar_import('transformers')
testar_import('FlagEmbedding', 'FlagEmbedding (BGE-M3)')
testar_import('faiss', 'faiss-cpu')
testar_import('opensearchpy', 'opensearch-py')
testar_import('langchain', 'langchain')
testar_import('openai')
testar_import('langfuse')
testar_import('umap', 'umap-learn')
testar_import('matplotlib')
testar_import('plotly')
testar_import('pandas')
testar_import('numpy')
testar_import('torch', 'pytorch', 'version')

# Exibe resultados em tabela
print(f'{'Biblioteca':<30} {'Status':<6} {'Versão'}')
print('-' * 60)
for nome, status, versao in resultados:
    print(f'{nome:<30} {status:<6} {versao}')

# Sumário
ok = sum(1 for _, s, _ in resultados if s == '✅')
total = len(resultados)
print(f'\n{'='*60}')
print(f'Resultado: {ok}/{total} bibliotecas instaladas corretamente')
if ok == total:
    print('✅ AMBIENTE COMPLETAMENTE CONFIGURADO!')
else:
    print(f'⚠️  {total-ok} biblioteca(s) com problema. Verifique a Célula 2.')
````

## 📦 CÉLULA 5 — Teste de GPU com PyTorch

**O que faz:** Verifica se o PyTorch consegue acessar a GPU e executa um cálculo simples para confirmar.

**Por que:** Os modelos de embedding e LLMs precisam da GPU para performance adequada.

````python
# Célula 5: Teste de GPU com PyTorch
import torch
import time

print('🎮 TESTE DE GPU COM PYTORCH')
print('=' * 60)
print(f'PyTorch versão: {torch.__version__}')
print(f'CUDA disponível: {torch.cuda.is_available()}')

if torch.cuda.is_available():
    device = torch.device('cuda')
    print(f'GPU detectada: {torch.cuda.get_device_name(0)}')
    print(f'Memória GPU: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB')
    
    # Benchmark simples: multiplicação de matrizes
    tamanho = 4096
    a = torch.rand(tamanho, tamanho, device=device)
    b = torch.rand(tamanho, tamanho, device=device)
    
    # Aquecimento
    _ = torch.matmul(a, b)
    torch.cuda.synchronize()
    
    # Medição
    inicio = time.time()
    for _ in range(10):
        c = torch.matmul(a, b)
    torch.cuda.synchronize()
    tempo = (time.time() - inicio) / 10
    
    print(f'\n✅ Benchmark GPU (matmul {tamanho}x{tamanho}): {tempo*1000:.1f} ms por operação')
    print('   GPU está funcionando corretamente!')
    
    # Limpeza
    del a, b, c
    torch.cuda.empty_cache()
else:
    device = torch.device('cpu')
    print('\n⚠️  GPU não disponível. Usando CPU.')
    print('   Os modelos funcionarão, mas serão mais lentos.')
    print('   Recomendação: Runtime → Alterar tipo de runtime → GPU T4')

print(f'\n📱 Dispositivo configurado: {device}')
DEVICE = device  # Variável global para uso nos outros labs
````

## 📦 CÉLULA 6 — Teste Rápido de Embedding (Smoke Test)

**O que faz:** Executa um teste mínimo de embedding para confirmar que sentence-transformers está funcionando corretamente com GPU.

**Por que:** Melhor descobrir problemas agora com um modelo pequeno do que durante o Lab 5 com BGE-M3.

````python
# Célula 6: Smoke test de embedding
from sentence_transformers import SentenceTransformer
import numpy as np
import time

print('🧪 SMOKE TEST: Sentence-Transformers')
print('=' * 60)

# Usa um modelo leve para o smoke test (paraphrase-MiniLM: apenas 22MB)
print('⏳ Carregando modelo leve para teste (paraphrase-MiniLM-L6-v2)...')
modelo_teste = SentenceTransformer('paraphrase-MiniLM-L6-v2', device=str(DEVICE))
print('✅ Modelo carregado!')

# Frases de teste com vocabulário jurídico
frases = [
    'O réu foi absolvido por insuficiência de provas.',
    'O acusado foi inocentado por falta de evidências.',  # Semanticamente similar
    'O tribunal condenou o servidor pelo crime de peculato.',  # Diferente
    'A vítima registrou boletim de ocorrência na delegacia.'
]

# Gera embeddings
inicio = time.time()
embeddings = modelo_teste.encode(frases, convert_to_numpy=True)
tempo = time.time() - inicio

print(f'\n📐 Dimensão dos embeddings: {embeddings.shape}')
print(f'⏱️  Tempo para {len(frases)} frases: {tempo:.2f}s')

# Calcula similaridade coseno entre as frases
from sklearn.metrics.pairwise import cosine_similarity

sim_matrix = cosine_similarity(embeddings)

print('\n📊 MATRIZ DE SIMILARIDADE COSENO:')
print('   (1.0 = idêntico, 0.0 = sem relação)')
print()
labels = ['Absolvido/provas', 'Inocentado/evidências', 'Peculato/STJ', 'BO/delegacia']
for i, label in enumerate(labels):
    sims = ' '.join(f'{sim_matrix[i,j]:.2f}' for j in range(len(frases)))
    print(f'  {label:<25} {sims}')

# Verifica que a similaridade entre frases 0 e 1 é alta
sim_01 = sim_matrix[0, 1]
if sim_01 > 0.7:
    print(f'\n✅ CORRETO! "Absolvido" ↔ "Inocentado": {sim_01:.2f} (alta similaridade)')
else:
    print(f'\n⚠️  Similaridade {sim_01:.2f} menor que esperado. Verifique instalação.')

print('\n✅ Smoke test concluído! Ambiente de NLP funcionando.')
````
